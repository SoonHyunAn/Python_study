```python
"""
#키보드 이벤트#
import pygame

pygame.init() # 초기화

#화면 크기 설장
screen_width = 480 #가로
screen_height = 640 #세로
screen = pygame.display.set_mode((screen_width,screen_height))

#화면 타이틀 설정
pygame.display.set_caption("Nado Game") #게임 이름

#배경 이미지 불러오기
background = pygame.image.load("H:/etc/Etc/Python/Untitled Folder/뉴버거와 똥강아지.png")

#스프라이트(캐릭터) 불러오기
character = pygame.image.load("H:/etc/Etc/Python/Untitled Folder/캐릭터 버거.png")
character_size = character.get_rect().size #캐릭터의 사각형 크기 얻어옴
character_width = character_size[0] #캐릭터 가로
character_height = character_size[1] #캐릭터 세로
character_x_pos = (screen_width / 2) - (character_width/2) #화면 가로의 절반 크기에 해당하는 곳에 위치
character_y_pos = screen_height - character_height  #화면 세로의 가장 아래에 해당하는 곳에 위치

# 이동할 죄표
to_x=0
to_y=0

#이벤트 루프
running = True #게임이 진행중인가?
while running:
    for event in pygame.event.get(): #이벤트가 지속해서 진행중인가?
        if event.type == pygame.QUIT: #창이 닫히는 이벤트가 발생하였는가?
            running = False #게임 종료
        if event.type == pygame.KEYDOWN: #방향키 입력
            if event.key == pygame.K_LEFT:
                to_x -=5
            elif event.key == pygame.K_RIGHT:
                to_x +=5
            elif event.key == pygame.K_UP:
                to_y -=5
            elif event.key == pygame.K_DOWN:
                to_y +=5
                
        if event.type == pygame.KEYUP: #방향키 입력이 없으면 동작을 멈춤
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                to_x=0
            elif event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                to_y=0
                
    character_x_pos += to_x
    character_y_pos += to_y
    
    #화면 밖으로 벗어날 때 경계값 설정
    if character_x_pos<0:
        character_x_pos=0
    elif character_x_pos >screen_width - character_width:
        character_x_pos =screen_width - character_width
    
    if character_y_pos<0:
        character_y_pos=0
    elif character_y_pos > screen_height - character_height:
        character_y_pos = screen_height - character_height
    
    
    screen.blit(background, (0, 0)) #맨 왼쪽 맨위 시작
    screen.blit(character, (character_x_pos, character_y_pos))
    pygame.display.update()

#pygame 종료
pygame.quit()
"""
```

```python
#정리! - 기본 게임 프레임 제작#

import pygame
#===================================================================-===================================
#기본 단계
pygame.init() # 초기화

#화면 크기 설장
screen_width = 480 #가로
screen_height = 640 #세로
screen = pygame.display.set_mode((screen_width,screen_height))

#화면 타이틀 설정
pygame.display.set_caption("") #게임 이름

#FPS
clock = pygame.time.Clock()

#===================================================================-===================================
# 사용자 게임 초기화 (배경, 이미지, 좌표, 폰트, 속도) - 아군 적군

#배경 이미지 불러오기
background = pygame.image.load("")

#스프라이트(캐릭터) 불러오기
character = pygame.image.load("")
character_size = character.get_rect().size #캐릭터의 사각형 크기 얻어옴
character_width = character_size[0] #캐릭터 가로
character_height = character_size[1] #캐릭터 세로
character_x_pos = (screen_width / 2) - (character_width/2) #화면 가로의 절반 크기에 해당하는 곳에 위치
character_y_pos = screen_height - character_height  #화면 세로의 가장 아래에 해당하는 곳에 위치

# 이동할 죄표
to_x=0
to_y=0

# 캐릭터 이동 속도
char_speed = 0.6

#폰트 정의
game_font = pygame.font.Font(None, 40)

#총 시간, 계산
total_time = 10
start_ticks = pygame.time.get_ticks() #시작 틱을 받아옴

#이벤트 루프
running = True #게임이 진행중인가?
while running:
    dt=clock.tick(60) #게임 화면의 초당 프레임 수
    
#===================================================================-===================================
# 이벤트 처리 (키보드, 마우스)
    for event in pygame.event.get(): 
        if event.type == pygame.QUIT: 
            running = False 
        if event.type == pygame.KEYDOWN: 
            if event.key == pygame.K_LEFT:
                to_x -=char_speed
            elif event.key == pygame.K_RIGHT:
                to_x +=char_speed
            elif event.key == pygame.K_UP:
                to_y -=char_speed
            elif event.key == pygame.K_DOWN:
                to_y +=char_speed
                
        if event.type == pygame.KEYUP: 
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                to_x=0
            elif event.key == pygame.K_UP or event.key == pygame.K_DOWN:
                to_y=0
                
#===================================================================-===================================
# 캐릭터 위치정보, 적 위치 정보, 충돌 정보             
    character_x_pos += to_x *dt
    character_y_pos += to_y *dt
    
    #화면 밖으로 벗어날 때 경계값 설정
    if character_x_pos<0:
        character_x_pos=0
    elif character_x_pos >screen_width - character_width:
        character_x_pos =screen_width - character_width
    
    if character_y_pos<0:
        character_y_pos=0
    elif character_y_pos > screen_height - character_height:
        character_y_pos = screen_height - character_height
    
    #적과 충돌 처리
    character_rect = character.get_rect()
    character_rect.left=character_x_pos
    character_rect.top = character_y_pos
    
    enemy_rect = enemy.get_rect()
    enemy_rect.left= enemy_x_pos
    enemy_rect.top = enemy_y_pos    
    
    if character_rect.colliderect(enemy_rect):
        print("충돌했어요")
        running = False

#===================================================================-===================================
# 화면에 표시      
    screen.blit(background, (0, 0)) #맨 왼쪽 맨위 시작
    screen.blit(character, (character_x_pos, character_y_pos))
    screen.blit(enemy, (enemy_x_pos, enemy_y_pos))
    
    #타이머 넣기 
    # 경과시간 ms단위
    elapsed_time = (pygame.time.get_ticks()-start_ticks)/1000 
    timer = game_font.render(str(int(total_time-elapsed_time)), True, (255,255,255))
    #출력글자, True 글자 색상
    screen.blit(timer, (10,10)) 
    if total_time-elapsed_time <= 0: #시간이 0이하일 경우 게임 종료
        running = False
   
    pygame.display.update()
    
#게임 종료 딜레이 ms단위로 약 2초 대기
pygame.time.delay(2000)
#pygame 종료
pygame.quit()
```
