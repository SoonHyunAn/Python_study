```python
"""
퀴즈 - 하늘에서 떨어지는 똥 피하기 게임을 만드시오

[게임조건]
1. 캐릭터는 화면 가장 아래에 위치, 좌우만 이동 가능
2. 똥은 화면 가장 위에서 떨어짐, x 좌표는 매번 랜덤으로 설정
3. 캐릭터가 똥을 피하면 다음 똥을 다시 떨어짐
4. 캐릭터가 똥과 충돌하면 게임 종료
5. FPA는 30으로 고정

[게임 이미지 크기]
1. 배경 640*480
2. 캐릭터 및 똥 70*70
"""
```

```python
#똥피하기 게임#

import pygame
import random as rand
#===================================================================-===================================
#기본 단계
pygame.init() # 초기화

#화면 크기 설장
screen_width = 640 #가로
screen_height = 480 #세로
screen = pygame.display.set_mode((screen_width,screen_height))

#화면 타이틀 설정
pygame.display.set_caption("똥강아지 똥피하지") #게임 이름

#FPS
clock = pygame.time.Clock()

#===================================================================-===================================
# 사용자 게임 초기화 (배경, 이미지, 좌표, 폰트, 속도) - 아군 적군

#배경 이미지 불러오기
background = pygame.image.load("H:/etc/Etc/Python/4. 파이썬 활용 - Game 만들기/뉴버거와 똥강아지.png")

#스프라이트(캐릭터) 불러오기
character = pygame.image.load("H:/etc/Etc/Python/4. 파이썬 활용 - Game 만들기/캐릭터 버거.png")
character_size = character.get_rect().size #캐릭터의 사각형 크기 얻어옴
character_width = character_size[0] #캐릭터 가로
character_height = character_size[1] #캐릭터 세로
character_x_pos = (screen_width / 2) - (character_width/2) #화면 가로의 절반 크기에 해당하는 곳에 위치
character_y_pos = screen_height - character_height  #화면 세로의 가장 아래에 해당하는 곳에 위치

#적 불러오기
enemy = pygame.image.load("H:/etc/Etc/Python/4. 파이썬 활용 - Game 만들기/enemy.jpg")
enemy_size = character.get_rect().size #캐릭터의 사각형 크기 얻어옴
enemy_width = character_size[0] #캐릭터 가로
enemy_height = character_size[1] #캐릭터 세로
enemy_x_pos = rand.randint(0, 480-enemy_width) #화면 가로의 절반 크기에 해당하는 곳에 위치
enemy_y_pos = 0 #화면 세로의 가장 아래에 해당하는 곳에 위치
enemy_speed = 10


# 이동할 죄표
to_x=0

# 캐릭터 이동 속도
char_speed = 0.6

#폰트 정의
game_font = pygame.font.Font(None, 40)

#총 시간, 계산
total_time = 100
start_ticks = pygame.time.get_ticks() #시작 틱을 받아옴

#이벤트 루프
running = True #게임이 진행중인가?
while running:
    dt=clock.tick(60) #게임 화면의 초당 프레임 수
    
#===================================================================-===================================
# 이벤트 처리 (키보드, 마우스)
    for event in pygame.event.get(): 
        if event.type == pygame.QUIT: 
            running = False 
            
        if event.type == pygame.KEYDOWN: 
            if event.key == pygame.K_LEFT:
                to_x -=char_speed
            elif event.key == pygame.K_RIGHT:
                to_x +=char_speed
                
        if event.type == pygame.KEYUP: 
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                to_x=0
                
#===================================================================-===================================
# 캐릭터 위치정보, 적 위치 정보, 충돌 정보             
    character_x_pos += to_x *dt

    
    #화면 밖으로 벗어날 때 경계값 설정
    if character_x_pos<0:
        character_x_pos=0
    elif character_x_pos >screen_width - character_width:
        character_x_pos =screen_width - character_width
    
    # 적 이동
    enemy_y_pos += enemy_speed
    
    if enemy_y_pos > screen_height:
        enemy_y_pos = 0
        enemy_x_pos = rand.randint(0, 480-enemy_width)
    
    #적과 충돌 처리
    character_rect = character.get_rect()
    character_rect.left=character_x_pos
    character_rect.top = character_y_pos
    
    enemy_rect = enemy.get_rect()
    enemy_rect.left= enemy_x_pos
    enemy_rect.top = enemy_y_pos    
    
    if character_rect.colliderect(enemy_rect):
        print("충돌했어요")
        running = False

#===================================================================-===================================
# 화면에 표시      
    screen.blit(background, (0, 0)) #맨 왼쪽 맨위 시작
    screen.blit(character, (character_x_pos, character_y_pos))
    screen.blit(enemy, (enemy_x_pos, enemy_y_pos))
    
    #타이머 넣기 
    # 경과시간 ms단위
    elapsed_time = (pygame.time.get_ticks()-start_ticks)/1000 
    timer = game_font.render(str(int(total_time-elapsed_time)), True, (255,255,255))
    #출력글자, True 글자 색상
    screen.blit(timer, (10,10)) 
    if total_time-elapsed_time <= 0: #시간이 0이하일 경우 게임 종료
        running = False
   
    pygame.display.update()
    
#게임 종료 딜레이 ms단위로 약 2초 대기
pygame.time.delay(2000)
#pygame 종료
pygame.quit()
```
