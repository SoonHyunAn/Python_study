```python
from tkinter import *

#기본 틀
root = Tk()
root.title("메모장")
root.geometry("640x480+100+300") 

root.resizable(False, False) 

#텍스트와 엔트리
text = Text(root, width=30, height=5)#텍스트 창 크기
text.pack()

text.insert(END,"글자를 입력하세요")

e= Entry(root,width=30) #엔터 입력 불가능
e.pack()

e.insert(END,"한줄만 입력하세요")

def btncmd():
    print(text.get("1.0",END)) # 1: 첫번쨰 라인, 0: 0번째 column 위치
    print(e.get())
    
    text.delete("1.0",END)
    e.delete(0, END)       #출력된 뒤 입력된 값을 모두 지움
    
btn =Button(root, text="클릭 -출력",command=btncmd)
btn.pack()    
    
root.mainloop()
```

```python
import tkinter.ttk as ttk
import time
from tkinter import *

#기본 틀
root = Tk()
root.title("메모장")
root.geometry("640x480+100+300") 

root.resizable(False, False) 

# 프로그래스 바

progressbar=ttk.Progressbar(root, maximum=100, mode="determinate") #indeterminate 비결정, 
progressbar.start(10) #10ms
progressbar.pack()


def btncmd1():
    progressbar.stop()
    
btn1 =Button(root, text="중지",command=btncmd1)
btn1.pack()

p_var2 = DoubleVar()
progressbar2=ttk.Progressbar(root, maximum=100, length = 150, variable = p_var2) 
progressbar2.pack()



def btncmd2():
    for i in range (101):
        time.sleep(0.01) #0.01초 대기 
        p_var2.set(i)
        progressbar2.update() # 시각적으로 프로그래스를 보여줌 (1cycle마다 리프레쉬)
        print(p_var2.get())
        
btn2 =Button(root, text="시작",command=btncmd2) # 화면에 나오는 값을 출력
btn2.pack()


root.mainloop()
```

```python
import tkinter.messagebox as msgbox
from tkinter import *


#기본 틀
root = Tk()
root.title("메모장")
root.geometry("640x480+100+300") 

root.resizable(False, False) 

# 메시지 박스 - 좌석 예약 서비스
def info():
    msgbox.showinfo("알림", "정상적으로 예매 완료되었습니다.")
def warn():  
    msgbox.showwarning("경고", "해당 좌석은 예매할 수 없습니다.")
def error():  
    msgbox.showerror("에러", "비정상 접근입니다.")
def okcancel():  
    msgbox.askokcancel("확인/취소", "해당 좌석은 유아용으로 성인이 앉을 수 없습니다. 예매하시겠습니까?")
def retrycancel():  
    response = msgbox.askretrycancel("재시도/취소", "일시적 오류가 발생했습니다. 다시 시도하시겠습니까?")
    if response == 1:
        print("예")
    else:
        print("아니요")
def yesno():  
    msgbox.askyesno("yes/no", "해당 방향은 역방향입니다. 예매하시겠습니까?")
def yesnocancel():  
    response = msgbox.askyesnocancel("yes/no/cancel", "예매 내역이 저장되지 않았습니다. 저장 하시겠습니까?")
    #네: 저장 후 종료(True, 1), 아니요: 저장하지 않고 종료(False, 0). 취소: 프로그램 종료 취소(None)
    if response == 1:
        print("예")
    elif response == 0:
        print("아니요")
    else:
        print("취소")
    
Button(root, command =info, text="알림").pack()
Button(root, command =warn, text="경고").pack()
Button(root, command =error, text="에러").pack()
Button(root, command =okcancel, text="확인").pack()
Button(root, command =retrycancel, text="재확인").pack()
Button(root, command =yesno, text="yesno").pack()
Button(root, command =yesnocancel, text="yesnocancel").pack()




root.mainloop()
```

```python
"""
Quiz: tkinter를 이용한 메모장 프로그램을 만드시오 (window 메모장과 비교)

1. title: 제목 없음
2. 메뉴: 파일, 편집, 서식, 보기, 도움말
3. 실제 메뉴 구현: 파일 메뉴 내에서 열기, 저장, 끝내기 3개만 처리
3-1. 열기: mynote.txt파일 내용 열어서 보여주기
3-2. 저장: mynote.txt파일에 현재 내용 저장하기
3-3. 끝내기: 프로그램 종료
4. 프로그램 시작시 본문은 비어있는 상태
5. 하단 statue 바는 필요없음
6. 프로그램 크기, 위치는 자유롭게 하되 크기 조정 가능해야 함
7. 본문 우측에 상하 스크롤 바 넣기
"""

from tkinter import *
import os


#기본 틀
root = Tk()
root.title("제목 없음 -Window 메모장")
root.geometry("640x480+100+300") 

file_name="mynote.txt"

def Openfile():
    if os.path.isfile(file_name): #파일 존재 여부
        with open(file_name, "r", encoding="utf8") as file: #r이면 read
            text.delete("1.0", END) #기존 내용 삭제
            text.insert(END, file.read()) #저장된 내용 불러오기
    
def Savefile():
    with open(file_name, "w", encoding="utf8") as file:  #w면 write
                  file.write(text.get("1.0", END)) #전체 저장

menu = Menu(root)
menu_file = Menu(menu, tearoff=0)
menu_file.add_command(label="열기", command=Openfile)
menu_file.add_command(label="저장", command=Savefile)
menu_file.add_command(label="끝내기", command = root.quit())
menu.add_cascade(label="파일", menu=menu_file)

menu.add_cascade(label="편집")
menu.add_cascade(label="서식")
menu.add_cascade(label="보기")
menu.add_cascade(label="도움말")

scrollbar = Scrollbar(root) 
scrollbar.pack(side="right", fill="y")

text = Text(root, yscrollcommand=scrollbar.set)
text.pack(fill="both", expand=True)

root.config(menu=menu)
scrollbar.config(command=text.yview) 
root.mainloop()
```
