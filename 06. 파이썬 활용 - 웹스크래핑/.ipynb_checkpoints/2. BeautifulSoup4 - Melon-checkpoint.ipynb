```python
# install BeautifulSoup4 (스크래핑 용), lxml (구문을 분석하는 파서)

import requests
from bs4 import BeautifulSoup

headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36"}
url="https://www.melon.com/"
res=requests.get(url, headers=headers)
res.raise_for_status()

soup = BeautifulSoup(res.text, "lxml") #모든 정보를 가지고 있음
# 잘 알때
print(soup.title)# html의 제목을 뽑아옴
print("\n")
print(soup.title.get_text())
print("\n")
#print(soup.div) # soup 객체에서 처음 발견되는 div 객체의 정보를 추출
#print("\n")
print(soup.div.attrs) # div의 속성정보를 추출
print("\n")
print(soup.div["id"]) # div의 id 속성 값 정보를 출력 (1단계 아래의 정보만 추출 가능)
print("\n")

# 잘 모를 때
rank1 = soup.find("div", attrs={"class":"rank_number nth1"}) # "div"는 생략 가능
print("================================= \n div-class-rank_number nth1: ",rank1.get_text(),"\n================================= \n")
rank2 = rank1.next_sibling.next_sibling
print("================================= \n next_sibling.next_sibling ", rank2.get_text(),"\n================================= \n")

rank3 = rank2.previous_sibling.previous_sibling
print("\n================================= \n previous_sibling.previous_sibling", rank3.get_text(),"\n================================= \n")

print("parent", rank1.parent.get_text())

rank4 = rank1.find_next_sibling("div")
print("\n================================= \n find_next_sibling",rank4.get_text(),"\n================================= \n")
rank5 = rank2.find_previous_sibling("div")
print("\n================================= \n find_previous_sibling",rank5.get_text(),"\n================================= \n")

music=soup.find("span", text="내일의 우리")
print(music)
```

```python
# install BeautifulSoup4 (스크래핑 용), lxml (구문을 분석하는 파서)

import requests
from bs4 import BeautifulSoup

headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36"}
url="https://www.melon.com/chart/index.htm"
res=requests.get(url, headers=headers)
res.raise_for_status()

soup = BeautifulSoup(res.text, "lxml") #모든 정보를 가지고 있음

musics = soup.find_all("a", attrs={"class": "ellipsis"}) 
# a라는 항목에 class=ellipsis이고, ellipsis에 들어간 값을 불러옴
for i in musics:
    print(i.get_text())

```
