영화 1 : 영화의 평점이 10/10 -> 5명이 평가

영화 2 : 영화의 평점이 8/10 -> 500명이 평가



v is the number of votes for the movie;

m is the minimum votes required to be listed in the chart;

R is the average rating of the movie; And

C is the mean vote across the whole report

```python
C = df2['vote_average'].mean() # 전체 영화의 평균
C
```

```python
m = df2['vote_count'].quantile(0.9) # 평가수가 상위 10% 영화의 평가수 기준
m
```

```python
q_movies = df2.copy().loc[df2['vote_count'] >= m] #df2를 복사해 사용
q_movies.shape #4800개중에 약 480개가 있음 (10%)
```

```python
q_movies['vote_count'].sort_values()
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v + m) * R) + (m / (m + v) * C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1) # 점수를 위한 새로운 열
q_movies.head(3)
```

```python
q_movies = q_movies.sort_values('score', ascending=False) #내림차순
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates()#중복 제거 (없긴함)
indices
```

```python
# 영화의 제목을 입력받으면 코사인 유사도를 통해서 가장 유사도가 높은 상위 10개의 영화 목록 반환
def get_recommendations(title, cosine_sim=cosine_sim):
    # 영화 제목을 통해서 전체 데이터 기준 그 영화의 index 값을 얻기
    idx = indices[title]
    
    # 코사인 유사도 매트릭스 (cosine_sim) 에서 idx 에 해당하는 데이터를 (idx, 유사도) 형태로 얻기
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # 코사인 유사도 기준으로 내림차순 정렬
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # 자기 자신을 제외([0]index)한 10개의 추천 영화를 슬라이싱
    sim_scores = sim_scores[1:11]
    
    # 추천 영화 목록 10개의 인덱스 정보 추출
    movie_indices = [i[0] for i in sim_scores]
    
    # 인덱스 정보를 통해 영화 제목 추출
    return df2['title'].iloc[movie_indices]
```

```python
# 감독 정보를 추출
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
df2['director'] = df2['crew'].apply(get_director)
df2['director']
```

```python
# 처음 3개의 데이터 중에서 name 에 해당하는 value 만 추출
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3: # name을 통해 얻어오는 정보가 3개 이상일 경우 3개까지만 표시
            names = names[:3]
        return names
    return []
```

```python
features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def clean_data(x): #띄어쓰기 제거, 모두 소문자로 변환
    if isinstance(x, list):
        return [str.lower(i.replace(' ', '')) for i in x]
    else:
        if isinstance(x, str): #혹시나 리스트 형식이 아닌 값이 있을 경우
            return str.lower(x.replace(' ', ''))
        else:
            return ''
```

```python
features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def create_soup(x): #제목을 제외한 나머지 열값을 띄어쓰기를 기준으로 합치는 작업
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
df2['soup']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english') #관사 확인
count_matrix = count.fit_transform(df2['soup']) # 각 키워드를 쪼개서 셈
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
# 제목 (title)을 제외하고 CountVectorizer를 통해 쪼개놓은 것과 비교한 
# Cosine 유사도를 통해 가장 가까운 값을 찾아냄 (출력값은 title)
cosine_sim2
```
